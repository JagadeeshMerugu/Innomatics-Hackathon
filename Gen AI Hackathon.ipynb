### 4. Which restaurant rating range generated the highest total revenue?

```python
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0–3.5', '3.6–4.0', '4.1–4.5', '4.6–5.0']

df['rating_range'] = pd.cut(
    df['rating'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
df.groupby('rating_range')['total_amount'].sum().sort_values(ascending=False)
```

### 6. Which cuisine has the lowest number of distinct restaurants but still contributes significant revenue?

```python
cuisine_data = df.groupby('cuisine').agg(
    distinct_restaurants = ('restaurant_id', 'nunique'),
    total_revenue = ('total_amount', 'sum')
)
cuisine_data
```

```python
median_revuenue = cuisine_data['total_revenue'].median()
median_revuenue
```

```python
results = cuisine_data[cuisine_data['total_revenue'] >= median_revuenue]
results
```

```python
results.sort_values('distinct_restaurants').head(1)
```

### 7. What percentage of total orders were placed by Gold members? (Rounded to nearest integer)

```python
total_orders = len(df)
gold_orders = len(df[df['membership'].str.lower() == 'gold'])
print(gold_orders)
percentage = round((gold_orders / total_orders) * 100)
print(percentage)
```

### 8. Which restaurant has the highest average order value but less than 20 total orders?

```python
restaurant_stats = df.groupby('restaurant_name').agg(
    total_orders = ('order_id', 'count'),
    avg_order_value = ('total_amount', 'mean')
)
restaurant_stats
```

```python
result = restaurant_stats[restaurant_stats['total_orders'] < 20].sort_values('avg_order_value', ascending=False)
result
```

```python
result.sort_values('total_orders', ascending=False).head(2)
```

### 9. Which combination contributes the highest revenue?

```python
options = [
    ('Gold', 'Indian'),
    ('Gold', 'Italian'),
    ('Regular', 'Indian'),
    ('Regular', 'Chinese')
]

results = []

for membership, cuisine in options:
    revenue = df[ 
        (df['membership'] == membership) &
        (df['cuisine'] == cuisine)
    ]['total_amount'].sum()
    
    results.append((membership+ " + "+ cuisine, revenue))

results
```
